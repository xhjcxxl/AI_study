```python
import re
# package cmd to send usb
class Package_Cmd:
    Max_Len = 512  # 最大长度
    cmd_List_Header = [0xff, 0x55, 0x56, 0x42]  # 命令 开头
    cmd_List_End = [0xfe]  # 命令 结尾
    
    def __init__(self):
        cmd_List = []  # 命令 内容 list
        cmd_List_Len = 0  # 命令 长度
        len_List = []  # 命令 长度 list
        cmd_List_Package = []  # 完整 命令 list

    # ------------------------------------------------
    # 初始化 将收的数据，转化为list，然后将判断数据是否符合要求
    # return 0 正常 | -1 错误
    def Init_Cmd(self, cmd_Str):
        self.cmd_Str = cmd_Str
        self.cmd_List = self.Str_To_List()  # 命令 转换
        self.cmd_List_Len = len(self.cmd_List)  # 命令 获取
        return self.Cmd_List_Chick()

    # ------------------------------------------------
    # 将cmd 字符串转化成列表
    # return list
    def Str_To_List(self):
        if len(self.cmd_Str) < 1:
            return 0
        self.cmd_Str = re.sub(r"(?<=\w)(?=(?:\w\w)+$)", " ", self.cmd_Str)  # 将无间隔的字符串 按照每2个字符 分割16进制字符 中间用空格补充
        cmd_List = self.cmd_Str.split(' ')  # 根据空格，将字符串直接转化为列表
        print(cmd_List)
        for i in range(len(cmd_List)):  # 循环列表，将列表中的字符，全部转化为16进制
            cmd_List[i] = int(cmd_List[i], 16)
        print(cmd_List)
        return cmd_List

    # ------------------------------------------------
    # 检查命令长度是否正确
    # return -1 长度错误 | 0 长度正确
    def Cmd_List_Chick(self):
        if (self.cmd_List_Len < 2) or (self.cmd_List_Len > Package_Cmd.Max_Len - 7):
            print("cmd len error!")
            return -1
        else:
            self.Get_Len_List()
            return 0

    # ------------------------------------------------
    # 获取命令长度
    def Get_Len_List(self):
        self.len_List = [(self.cmd_List_Len >> 8) & 0xff, self.cmd_List_Len & 0xff]

    # ------------------------------------------------
    # 整合命令
    def Get_Package_Cmd(self):
        self.cmd_List_Package = Package_Cmd.cmd_List_Header + self.len_List + self.cmd_List + Package_Cmd.cmd_List_End
        return self.cmd_List_Package
p1 = Package_Cmd()
ret = p1.Init_Cmd("C10101")
print(ret)
if ret == -1:
    print("cmd error")
else:
    print("cmd ok")
    print(p1.Get_Package_Cmd())
```
