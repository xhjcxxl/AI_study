```python
from numpy import *
import operator
import matplotlib
import matplotlib.pyplot as plt
from os import listdir
```

```python
def img2vector(filename):#文件数据为 32X32（图片 转 向量，矩阵）
    returnVect = zeros((1,1024))#1024 = 32X32；向量为1X1024
    fr = open(filename)
    for i in range(32):#32行
        lineStr = fr.readline()#读取一行数据
        for j in range(32):#一行 32 列
            returnVect[0, 32*i + j] = int(lineStr[j])#将列里面的所有数据都放到数组中
    return returnVect#返回矩阵
```

```python
def kNNClassify(newInput, dataSet, labels, k): # kNN算法分类 第一个为输入数据 第二个为训练数据  第三个为标签值 第四个为k的范围
    numSamples = dataSet.shape[0] # shape[0] stands for the num of row  
    ## step 1: calculate Euclidean distance  
    # tile(A, reps): Construct an array by repeating A reps times  
    # the following copy numSamples rows for dataSet  
    diff = tile(newInput, (numSamples, 1)) - dataSet # Subtract element-wise  
    squaredDiff = diff ** 2 # squared for the subtract  
    squaredDist = sum(squaredDiff, axis = 1) # sum is performed by row  
    distance = squaredDist ** 0.5  
  
    ## step 2: sort the distance  
    # argsort() returns the indices that would sort an array in a ascending order  
    sortedDistIndices = argsort(distance)  
  
    classCount = {} # define a dictionary (can be append element)  
    for i in range(k):  
        ## step 3: choose the min k distance  
        voteLabel = labels[sortedDistIndices[i]]  
  
        ## step 4: count the times labels occur  
        # when the key voteLabel is not in dictionary classCount, get()  
        # will return 0  
        classCount[voteLabel] = classCount.get(voteLabel, 0) + 1  
  
    ## step 5: the max voted class will return  
    maxCount = 0  
    for key, value in classCount.items():  
        if value > maxCount:  
            maxCount = value  
            maxIndex = key  
  
    return maxIndex
```

```python
def handwritingClassTest():#手写 测试 算法
    hwLabels = []
    trainingFileList = listdir('digits/trainingDigits') #返回指定文件夹下的文件、文件夹名字，以字母排序，不包括 \. \..
    m = len(trainingFileList)#获取文件个数
    trainingMat = zeros((m, 1024))#设置矩阵 多少张图片，就多少张 m X 1024
    for i in range(m):#循环数据
        fileNameStr = trainingFileList[i]#获取对应文件名字
        fileStr = fileNameStr.split('.')[0]#文件内容，以点为间隔，划分为数组，然后取[0] ，即第一个
        classNumStr = int(fileStr.split('_')[0])#数据格式转换
        hwLabels.append(classNumStr)#追加到列表末尾
        trainingMat[i,:] = img2vector('digits/trainingDigits/%s' % fileNameStr)#全部转换
    testFileList = listdir('digits/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):#前面部分为训练， 后面部分为进行分类预测
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest  = img2vector('digits/testDigits/%s' % fileNameStr)
        classifierResult = kNNClassify(vectorUnderTest, trainingMat, hwLabels, 3)#分类结果，真实结果
        print("the classifier came back with : %d, the real answer is : %d" % (classifierResult, classNumStr))
        if (classifierResult != classNumStr)：#计算错误个数
            errorCount += 1.0
    print("\n the total number of errors is : %d " % errorCount)#打印错误个数
    print("\n the total errors rate is : %f" % (errorCount/float(mTest)))#算出错误率
        
```
