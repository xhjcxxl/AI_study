```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.array(Image.open('longkui.jpg'))#array
plt.figure('longkui')#
plt.imshow(img)#
plt.axis('off')#
plt.show()#
```

```python
print(img.shape)#
print(img.dtype)#
print(img.size)#
print(type(img))#
```

```python
rows, cols, dims = img.shape#
print(rows, cols, dims)
```

```python
for i in range(100):#
    x=np.random.randint(0, rows)# 0-rows
    y=np.random.randint(0, cols)# 0-cols
    img[x, y, :] = 255
plt.imshow(img)
plt.show()
```
