```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.array(Image.open('longkui.jpg'))#打开图片并转化为array格式
plt.figure('longkui')#配置名字
plt.imshow(img)#显示图片
plt.axis('off')#关掉轴
plt.show()#搭载显示
```

```python
print(img.shape)#图片维度
print(img.dtype)#图片的数据格式
print(img.size)#图片的长度
print(type(img))#图片的存储类型
```

```python
rows, cols, dims = img.shape#直接将维度分为了三个部分
print(rows, cols, dims)
```

```python
for i in range(100):#随机点的个数
    x=np.random.randint(0, rows)#随机数 0-rows之间
    y=np.random.randint(0, cols)#随机数 0-cols之间
    img[x, y, :] = 255
plt.imshow(img)
plt.show()
```
